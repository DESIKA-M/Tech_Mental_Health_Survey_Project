# **Project Summary -**

Mental health in the workplace has become a critical concern, especially in fast-paced and high-pressure industries such as technology. This project focuses on analyzing a workplace mental health survey dataset to understand the factors influencing employees’ decisions to seek mental health treatment and to identify organizational practices that support or hinder mental well-being.

The dataset contains demographic details, work environment attributes, company policies, and perceptions related to mental and physical health. The primary objective of this project is to uncover meaningful patterns, correlations, and insights that can help organizations improve mental health awareness, policies, and support systems. The target variable for analysis is treatment, which indicates whether an individual has sought mental health treatment.

The project follows a structured UBM (Univariate, Bivariate, and Multivariate) analysis approach.

Univariate analysis was conducted to understand individual variable distributions such as age, gender, treatment status, work interference, and company size. This helped in identifying dominant categories, trends, and potential data imbalances.

Bivariate analysis explored relationships between treatment and other categorical and numerical variables such as family history, remote work, tech company status, benefits availability, supervisor support, and number of employees. Visualizations such as count plots, bar charts, and pie charts were used to clearly represent these relationships.

Multivariate analysis, including pair plots and correlation heatmaps, was performed on encoded features to examine interactions between multiple variables simultaneously and to identify key drivers influencing treatment-seeking behavior.

Before analysis, extensive data preprocessing and feature engineering were carried out. Missing values were handled appropriately, binary categorical variables were label encoded, and nominal variables were transformed using one-hot encoding. This ensured that the dataset was machine-learning ready and suitable for correlation analysis. The final encoded dataset was saved for future predictive modeling tasks.

Key insights from the analysis revealed that employees with a family history of mental illness, those working in larger organizations, and those with access to mental health benefits and supportive work cultures were more likely to seek treatment. Conversely, lack of anonymity, fear of negative consequences, and unsupportive supervisors were identified as barriers to treatment. The correlation heatmap highlighted that work environment factors and organizational policies had a stronger influence on treatment decisions than demographics alone.

Overall, this project demonstrates how data-driven insights can be leveraged to understand workplace mental health dynamics. The findings can assist organizations in designing inclusive mental health policies, improving employee support systems, and fostering a culture where mental well-being is treated with the same importance as physical health.

## 3. ***Data Wrangling***

### Data Wrangling Code

DATA CLEANING

````python
# AGE OUTLIERS REMOVAL
#copy is used to avoid SettingWithCopyWarning while filtering
df1 = df[(df['Age']>=18) & (df['Age']<=100)].copy()

# GENDER CLEANING
def gender_clean(Gender):
    gender = str(Gender).lower().strip()
    
    male_terms = ['male', 'm ', 'male-ish', 'maile', 'mal', 'male (cis)', 'cis male', 'male leaning androgynous', 'sissymale', 'neuter', 'something kinda male?', 'a little about you', 'ostensibly male', 'man', 'make']
    female_terms = ['female', 'f', 'woman', 'femina', 'female (cis)', 'cis female', 'femail', 'cis-female/femme', 'female (trans)', 'trans woman']
    
    if gender in male_terms or (gender.startswith('m') and len(gender)<3 and gender != 'me'):
        return 'Male'
    elif gender in female_terms or (gender.startswith('f') and len(gender)<3):
        return 'Female'
    else:  
        return 'Other/Trans'
    
df1['Gender'] = df1['Gender'].apply(gender_clean)

# Display the value counts of the cleaned gender column
df1['Gender'].value_counts()

# Dropping comments column
df1.drop (['comments'],axis =1, inplace=True)

# Dropping Timestamp column
df1.drop (['Timestamp'],axis =1, inplace=True)

# Filling missing values in 'self_employed' with mode
mode_self_employed = df1['self_employed'].mode()[0]
df1['self_employed'].fillna(mode_self_employed, inplace=True)

# Filling missing values in 'work_interfere' with 'Not Stated'
df1['work_interfere'].fillna('Not Stated', inplace=True)

# Filling missing values in 'state' with 'NA'
df1['state'].fillna('NA', inplace=True)

df1.to_csv("fully_cleaned_survey_data.csv", index=False)    

````

ENCODING FEATURES

````python
# ENCODING CATEGORICAL VARIABLES

x = df1.drop('treatment', axis=1)
y = df1['treatment']

# Columns to be label encoded
label_encode_cols = ['self_employed', 'family_history', 'remote_work', 'tech_company',  
                     'obs_consequence']

# Columns to be one-hot encoded
one_hot_cols = x.select_dtypes(include =['object']).columns.tolist()

# Remove label encoded columns from one-hot list
one_hot_cols =[col for col in one_hot_cols if col not in label_encode_cols]

# Label Encoding
for col in label_encode_cols :
    x[col] = x[col].replace({'Yes':1,'No':0})

# Encoding target variable
y_encoded = y.replace ({'Yes':1,'No':0})    

# One-Hot Encoding
x_encoded = pd.get_dummies(x, columns=one_hot_cols, drop_first=True)

# Combining encoded features and target variable
x_final = pd.concat([x_encoded,y_encoded],axis=1)

# Saving the final encoded dataframe to a CSV file
encoded_csv = x_final.to_csv("final_encoded_labeled_survey_data.csv", index=False)
````

### What all manipulations have you done and insights you found?

**Manipulations Performed**

1. **Feature–target separation:** Split the dataset into predictors (`x`) and target variable (`treatment`) to prepare for supervised learning.
2. **Encoding strategy definition:** Identified binary categorical columns for **label encoding** and multi-category object columns for **one-hot encoding**.
3. **Label encoding:** Converted binary responses (`Yes/No`) into numerical form (1/0) for model compatibility and interpretability.
4. **Target encoding:** Encoded the target variable (`treatment`) into binary numeric form (1 = Yes, 0 = No).
5. **One-hot encoding:** Transformed remaining categorical features into dummy variables, using `drop_first=True` to avoid multicollinearity.
6. **Feature consolidation:** Combined all encoded predictors with the encoded target into a single modeling-ready dataset.
7. **Data export:** Saved the final processed dataset as a CSV file for reuse in modeling and deployment pipelines.

**Insights Gained**

* Binary variables (e.g., family history, remote work, tech company) are now directly interpretable and suitable for most ML algorithms.
* One-hot encoding preserves categorical information without imposing artificial order, improving model fairness and accuracy.
* Dropping the first dummy level reduces redundancy and stabilizes linear and tree-based models.
* The final dataset is fully numeric, consistent, and scalable—ready for correlation analysis, feature importance evaluation, and predictive modeling.
